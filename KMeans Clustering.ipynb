# K Means Clustering


---

## üí° First, What is a **Cluster**?

A **cluster** means a **group of similar things**.

Imagine you're organizing your wardrobe. You put:

* All shirts together in one section,
* Pants in another section,
* Jackets in a third.

These **sections** are your **clusters** ‚Äî they contain similar items.

In machine learning, a **cluster** is a group of **similar data points**.

---

## ü§ñ What is K-Means Clustering?

**K-Means Clustering** is an **unsupervised machine learning algorithm** used to **group data into K clusters** ‚Äî where **K is a number you choose**.

The main idea is:

> ‚ÄúPut similar data points into one group, and different data points into different groups.‚Äù

### üîπ Why "K" Means?

* **K** is just a number.
* It tells the algorithm how many **clusters** (groups) you want to divide your data into.

For example:

* If **K = 2**, divide the data into 2 groups.
* If **K = 3**, divide into 3 groups, and so on.

---

## üç≠ Simple Real-Life Example

Suppose you have 100 candies üç¨ of different **sizes** and **colors**. You don‚Äôt know anything about them.

Now you want to **group similar candies together**:

* Maybe by size (big vs. small),
* Or color (red, green, yellow).

You tell the machine:

> "I want to group them into **3 types**."

The machine doesn‚Äôt know what type they are, but it will:

1. Look at the candies' features (like size and color),
2. Try to divide them into **3 groups** that are most similar,
3. Each group = one **cluster**.

That‚Äôs exactly what **K-Means** does.

---

## üî¢ How K-Means Works (Step-by-Step)

Let‚Äôs say you have 2 features: **Age** and **Income**.

You want to find **customer groups**. Here's how K-Means works:

---

### üîπ Step 1: Choose K (number of clusters)

* Let‚Äôs say **K = 2** (we want to divide customers into 2 groups).

---

### üîπ Step 2: Randomly select K points (called **centroids**)

* These are like imaginary "centers" of each cluster.
* Initially, they are **random**.

---

### üîπ Step 3: Assign each data point to the **nearest centroid**

* Each customer is assigned to the cluster with the **closest centroid** (using distance formula).
* This makes temporary groups.

---

### üîπ Step 4: Recalculate the centroid of each group

* The algorithm now finds the **average point (center)** of each group and moves the centroid there.

---

### üîπ Step 5: Repeat steps 3 and 4

* Points are reassigned to new centroids.
* Centroids are updated again.

This continues until:

* The points stop changing clusters
* Or centroids don‚Äôt move anymore.

This is called **convergence**.

---

## üìà What Does K-Means Give You?

* It labels each data point with a **cluster number** (0, 1, 2... up to K-1)
* You can now:

  * Visualize the clusters
  * Understand customer segments
  * Group products
  * Detect outliers (data that doesn't fit in any group)

---

## üîç Where is K-Means Used?

| Use Case                     | Example                                      |
| ---------------------------- | -------------------------------------------- |
| Customer Segmentation        | Group customers based on behavior            |
| Market Segmentation          | Divide buyers by region, interest, or budget |
| Image Compression            | Reduce number of colors in an image          |
| Pattern Recognition          | Group similar shapes, texts, etc.            |
| Document Clustering          | Group similar articles or news               |
| Anomaly Detection (indirect) | Find outliers that don‚Äôt belong to any group |

---

## üìâ What are the Limitations?

| Problem                 | Explanation                                          |
| ----------------------- | ---------------------------------------------------- |
| You must choose K       | You have to **know or guess** the number of clusters |
| Sensitive to outliers   | One odd point can shift centroids a lot              |
| Assumes circular shapes | Not good for complex shaped clusters                 |
| Random Initialization   | May give different results each time                 |

---

## üß™ Code Example (2D):

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Sample data
data = pd.DataFrame({
    'Age': [20, 21, 19, 35, 36, 37],
    'Income': [20000, 21000, 19000, 60000, 62000, 61000]
})

# Apply KMeans
model = KMeans(n_clusters=2)
model.fit(data)

# Add the cluster labels
data['Cluster'] = model.labels_

# Visualize
plt.scatter(data['Age'], data['Income'], c=data['Cluster'], cmap='rainbow')
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("KMeans Clustering")
plt.show()
```

## ü§î What is the Problem?

In **K-Means Clustering**, you must choose a number `K` (how many clusters or groups you want).

But the question is:

> **"How do I choose the right number of clusters?"**
> **K = 2? 3? 5? 10?** ü§∑

You can‚Äôt just guess! That‚Äôs where **Elbow Method** comes in.

---

## üéØ What is Elbow Method?

**Elbow Method** is a technique to help you **find the best K (number of clusters)** by checking **how well K-Means fits the data** for different K values.

We calculate something called **WCSS** for each K, and **plot it on a graph**.

Then, we **look for a point where the graph "bends like an elbow"**.
That point gives us the **best K value**.

---

## üìâ What is WCSS?

**WCSS** = *Within-Cluster Sum of Squares*

In simple terms:

> WCSS tells us **how tightly the data points are grouped within each cluster**.

### Easy Example:

* If WCSS is **low**, that means clusters are **tight and good**.
* If WCSS is **high**, clusters are **loose and bad**.

We want **lower WCSS**.

---

## üìà Elbow Method Graph

You plot:

* **X-axis** = Number of clusters `K`
* **Y-axis** = WCSS for each K

The graph will show that as you **increase K**, the **WCSS decreases** (clusters get better).

But after a point, the decrease becomes very small.

That **bend or elbow point** = **Best K**

---

## üî¢ Step-by-Step Code Example:

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample data
data = pd.DataFrame({
    'Age': [22, 25, 47, 52, 46, 56, 26, 30, 45, 55],
    'Income': [15000, 29000, 48000, 60000, 52000, 61000, 30000, 34000, 50000, 62000]
})

wcss = []

# Try K from 1 to 10
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)  # inertia_ = WCSS

# Plot the elbow graph
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS')
plt.title('Elbow Method For Optimal K')
plt.grid(True)
plt.show()
```

---

## üîç What to look for in the graph?

* The point where the graph **starts flattening** or **bends sharply like an elbow** is the best `K`.
* That‚Äôs where **adding more clusters doesn‚Äôt improve much**.

---

## üß† In Simple Words:

| Term      | Meaning                                                                 |
| --------- | ----------------------------------------------------------------------- |
| **WCSS**  | Total distance between points and their cluster center (lower = better) |
| **Elbow** | Point where adding more clusters stops helping much                     |
| **Goal**  | Find best K (clusters) without overdoing it                             |

---

## ‚úÖ Summary:

* **Elbow Method** = Way to find best `K` for K-Means
* **WCSS** = How compact clusters are
* **Best K** = Where graph bends (elbow)
* **Why important?** = So you don‚Äôt under-cluster or over-cluster your data

---

## üéØ What is a **Centroid** in KMeans?

A **centroid** is the **center point of a cluster**.

> It's **not** always a real data point ‚Äî it's a **calculated average point** that **represents the center of all points in the cluster**.

Think of it like this:

* Suppose you and your 4 friends are standing on a ground in different places.
* The **middle spot between all of you** is the **centroid**.

In **KMeans**, each cluster has **one centroid**.

---

## üì¶ Example:

Let‚Äôs say we have a cluster of 3 points (in 2D):

| X | Y |
| - | - |
| 1 | 2 |
| 2 | 3 |
| 3 | 4 |

Then, the **centroid (center)** is calculated as:

* **X mean** = (1 + 2 + 3) / 3 = **2**
* **Y mean** = (2 + 3 + 4) / 3 = **3**

So, the **centroid = (2, 3)**

This point may or may not exist in the actual data. It‚Äôs a **virtual center**.

---

## üîÅ How a point becomes a centroid in KMeans?

Here‚Äôs how the algorithm decides centroids:

### üîπ Step 1: Randomly pick `K` points as initial centroids

These can be:

* Random data points, or
* Random locations in feature space

### üîπ Step 2: Assign all other points to the nearest centroid

This creates clusters.

### üîπ Step 3: Calculate **new centroid** of each cluster

* Take **mean of all points** in each cluster
* That mean = **new centroid**

### üîπ Step 4: Repeat steps 2 and 3

* Until centroids **stop changing** (converge)

---

## üîÑ Centroid Update Example:

Imagine 5 points are grouped in a cluster:

| Age | Income |
| --- | ------ |
| 20  | 20000  |
| 22  | 22000  |
| 25  | 25000  |
| 23  | 24000  |
| 21  | 23000  |

Centroid = Mean of Age & Income

* Age mean = (20+22+25+23+21)/5 = 22.2
* Income mean = (20000+22000+25000+24000+23000)/5 = 22800

So centroid = (22.2, 22800)

---

## üìå Summary:

| Term                    | Meaning                                                                 |
| ----------------------- | ----------------------------------------------------------------------- |
| **Centroid**            | The average center of a group of points in a cluster                    |
| **How chosen**          | Initially random, then updated as **mean of all points in the cluster** |
| **Is it a real point?** | Not always. It‚Äôs often an imaginary/virtual average location            |

---

## üß† One-Line Definition:

> **Centroid** is the center of a cluster, calculated as the **average of all data points** in that cluster.
