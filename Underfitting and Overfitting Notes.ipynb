# ‚úÖ What is Underfitting?

```python
üî• Causes of Underfitting:
| Cause                             | Example                                                 |
| --------------------------------- | ------------------------------------------------------- |
| Model too simple                  | Using Linear Regression when data is complex/non-linear |
| Not enough features               | Ignoring important variables like location              |
| Too much regularization (penalty) | High values of Ridge/Lasso Œ±                            |
| Too few training epochs           | In deep learning, training stopped early                |
| Wrong algorithm                   | Using KNN with large `k` (very smooth)                  |

üõ†Ô∏è How to Fix Underfitting:
Use a more complex model (e.g., from Linear to Polynomial Regression)
Add more relevant features
Reduce regularization (lower alpha in Ridge/Lasso)
Train longer if using neural nets
```

# ‚úÖ What is Overfitting?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Step 1: Create sample curved data (X vs Y)
np.random.seed(0)
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])  # curve + noise

# Step 2: Create models with different complexity
models = [
    ("Underfit (Linear)", make_pipeline(PolynomialFeatures(1), LinearRegression())),
    ("Good Fit (Degree 3)", make_pipeline(PolynomialFeatures(3), LinearRegression())),
    ("Overfit (Degree 15)", make_pipeline(PolynomialFeatures(15), LinearRegression()))
]

# Step 3: Plot results
plt.figure(figsize=(15, 4))

for i, (title, model) in enumerate(models):
    model.fit(X, y)
    X_test = np.linspace(0, 5, 100).reshape(-1, 1)
    y_pred = model.predict(X_test)
    
    plt.subplot(1, 3, i + 1)
    plt.scatter(X, y, color='blue', label='Data')
    plt.plot(X_test, y_pred, color='red', label='Model')
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("y")
    plt.legend()

plt.tight_layout()
plt.show()
```
